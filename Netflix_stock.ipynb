```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from datetime import datetime
```

```python
df = pd.read_csv("Netflix.csv")
df
```

```python
sns.set(rc={'figure.figsize':(10,5)})
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df = df.set_index('Date')
df.head()
```

VOLUME OF STOCK TRADED

```python
sns.lineplot(x = df.index, y=df['Volume'], label = 'Volume')
plt.title('Volume of Stock versus Time')
```

NETFLIX STOCK PRICE - HIGH, LOW, OPEN, CLOSE

```python
df.plot(y = ['High','Low', 'Open'], title = 'Netflix Stock Price')
```

NETFLIX STOCK PRICE - DAY, MONTHS AND YEAR WISE

```python
import matplotlib.pyplot as plt

fig, (ax1, ax2, ax3) = plt.subplots(3, figsize=(15, 10))

# Group by day
df.groupby(df.index.day)['Volume'].mean().plot(ax=ax1)
ax1.set_xlabel('Day')
ax1.set_ylabel('Average Volume')

# Group by month
df.groupby(df.index.month)['Volume'].mean().plot(ax=ax2)
ax2.set_xlabel('Month')
ax2.set_ylabel('Average Volume')

# Group by year
df.groupby(df.index.year)['Volume'].mean().plot(ax=ax3)
ax3.set_xlabel('Year')
ax3.set_ylabel('Average Volume')

plt.tight_layout()
plt.show()
```

HIGH AND LOW VALUES STOCK PER PERIOD OF TIME

```python
fig,axes = plt.subplots(nrows =1, ncols = 2, sharex = True, figsize = (12,5))
fig.suptitle('High and Low values stock per period of time', fontsize = 18)
sns.lineplot(ax = axes[0], y = df['High'], x = df.index, color = 'green')
sns.lineplot(ax = axes[1], y = df['Low'], x = df.index, color = 'red')
```

TOP-5 DATES WITH HIGHEST STOCK PRICE

```python
a = df.sort_values(by = 'High', ascending = False)
a.head(5)
```

TOP-5 DATES WITH LOWEST STOCK PRICE

```python
a = df.sort_values(by = 'Low', ascending = True)
a.head(5)
```
